3.β. Υπάρχουν απουσιάζουσες τιμές; Γράψτε κατάλληλο κώδικα ώστε να χειριστείτε αυτές τις τιμές.

**Προσέγγιση: Target Oriented Filling αντί SimpleImputer**

Ο παρακάτω κώδικας συμπληρώνει τα κενά με βάση τη σχέση τους με τον στόχο 'RainTomorrow'. Ειδικότερα, για κάθε κλάση του στόχου (0 ή 1), τα κενά συμπληρώνονται με την πιο συχνή εμφάνισή τους - εάν είναι κατηγορικά - ή με τον μέσο όρο τους - εάν είναι αριθμητικά. Έτσι, δεν χρησιμοποιούμε απλά έναν SimpleImputer, αλλά η συμπλήρωση των απουσιαζουσών τιμών γίνεται βέλτιστα (υπάρχει όμως κίνδυνος για overfitting;).

```python
# Want to be sure that the original df is intact
try:
  df = pd.read_csv("train-val.csv")
  print("File loaded successfully.")

except FileNotFoundError:
  print("Error: 'train-val.csv' not found. Please ensure the file is in the correct location.")
except Exception as e:
  print(f"An error occurred: {e}")

df_imputed = df.copy()

# Fill missing values based on the mean or median of each feature grouped by 'RainTomorrow'
# Check for missing values again
print("\nMissing values per feature before imputation:")
print(df.isnull().sum())

# List of features with missing values
features_with_na = df.columns[df.isna().any()].tolist()

# Fill missing values with mean/median grouped by 'RainTomorrow'
for feature in features_with_na:
    # Check if the feature is numerical or categorical
    if df_imputed[feature].dtype in ['float64', 'int64']:
        # Use the mean for numerical features
        df_imputed[feature] = df_imputed.groupby('RainTomorrow')[feature].transform(lambda x: x.fillna(x.mean()))
    else:
        # For categorical features, use the most frequent (mode) value within each RainTomorrow group
        df_imputed[feature] = df_imputed.groupby('RainTomorrow')[feature].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))

# Verification - Output
print("\nMissing values per feature after imputation:")
print(df_imputed.isnull().sum())
```

3.γ. Γράψτε κώδικα για την κατάλληλη μετατροπή των κατηγορικών μεταβλητών ώστε να μπορούν να τους διαχειριστούν οι ταξινομητές που θα χρησιμοποιήσετε.

**Προσέγγιση: Target Encoding**

Η κωδικοποίηση αυτή λαμβάνει υπόψιν της τη σχέση με τον στόχο! Ειδικότερα, μετατρέπουμε τις κατηγορικές μεταβλητές σε αριθμητικές, βασιζόμενοι στη σχέση τους με τη μεταβλητή-στόχο (RainTomorrow). Αυτό γίνεται μέσω του υπολογισμού της μέσης τιμής της μεταβλητής-στόχου, για κάθε μοναδική τιμή της κατηγορικής μεταβλητής. Έτσι, παράγονται αριθμητικά δεδομένα τα οποία μπορούμε να χρησιμοποιήσουμε στα μοντέλα μας για τις προβλέψεις (π.χ. μέσω του correlation).

```python
# Target Feature to predict
target = 'RainTomorrow'

# List of features to encode (excluding target and ID columns)
features_to_encode = df_imputed.columns.drop([target, 'ID', 'Date'])
df_encoded = df_imputed.copy()

# Apply target encoding for each feature
for feature in features_to_encode:
    # Calculate the mean of 'RainTomorrow' for each unique value in the feature
    means = df_imputed.groupby(feature)[target].mean()
    # Map these means back to the DataFrame
    df_encoded[feature] = df_imputed[feature].map(means)

# Display the first few rows of the encoded DataFrame to verify
print("\nHead of the Encoded Dataframe:")
df_encoded.head()
```

Ακολούθως, συνενώνουμε τα χαρακτηριστικά που έχουν βαθμό συσχέτισης πάνω από ένα καθορισμένο threshold, ως 'κοινά'. Χρησιμοποιούμε ως βάρη της συνένωσης τον βαθμό συσχέτισής τους με τον στόχο!

```python
import pandas as pd
import numpy as np

# Copy the original dataframe and remove the 'Date' column
df_final2 = df_final.copy()
del df_final2['Date']
correlation_matrix = df_final2.corr()

# Threshold for high correlation between features
threshold = 0.75

# Get the list of highly correlated feature pairs
high_corr_pairs = [
    (col1, col2) for col1 in correlation_matrix.columns for col2 in correlation_matrix.columns
    if col1 != col2 and abs(correlation_matrix[col1][col2]) > threshold
]

# Keep track of processed features to avoid redundancy
processed_features = set()

# Merge highly correlated features
for col1, col2 in high_corr_pairs:
    if col1 in processed_features or col2 in processed_features:
        continue  # Skip already processed features

    # Calculate correlation with the target
    corr_with_target = df_final2.corr()['RainTomorrow']
    weight_col1 = abs(corr_with_target.get(col1, 0))
    weight_col2 = abs(corr_with_target.get(col2, 0))

    # Normalize weights
    total_weight = weight_col1 + weight_col2
    if total_weight == 0:
        continue  # Skip if both weights are zero
    weight_col1 /= total_weight
    weight_col2 /= total_weight

    # Create a new merged feature
    merged_feature_name = f"{col1}_{col2}_merged"
    df_final2[merged_feature_name] = weight_col1 * df_final2[col1] + weight_col2 * df_final2[col2]

    # Mark the original features as processed
    processed_features.update([col1, col2])

# Drop the original highly correlated features
features_to_drop = {col for col1, col2 in high_corr_pairs for col in (col1, col2)}
df_final2 = df_final2.drop(columns=list(features_to_drop.intersection(processed_features)))

# Display the merged Dataframe
print("Head of Merged Dataframe:")
print(df_final2.head())

# -- Let's set a correlation threshold (feaures with >= min thrsh_corr) --
correlation_threshold = 0.5

# Get the correlation values for the target variable (RainTomorrow)
correlation_matrix = df_final2.corr()
target_correlation = correlation_matrix['RainTomorrow'].abs()
selected_features = target_correlation[target_correlation >= correlation_threshold].index.tolist()
selected_features.remove('RainTomorrow') # Remove self (corr == 1)

print(f"\n\nFeatures with correlation above the set threshold ({correlation_threshold}):", selected_features)
```

Τέλος, κάνουμε scale όλα τα χαρακτηριστικά, ανάλογα με τον βαθμό συσχέτισής τους με τον στόχο (είναι όμως σωστή τακτική, αφού το corr() μετράει μόνο γραμμικές σχέσεις;).

```python
df_formodels = df_final2.copy()
correlations = df_final2.corr()['RainTomorrow'].drop('RainTomorrow')

# Apply correlation-based scaling to the selected features
for feature in correlations.index:
    weight = abs(correlations[feature])
    df_formodels[feature] *= weight

print("Correlation-weighted scaling completed successfully.")
df_formodels.head()
```

 4.α. Εκπαιδεύστε (fit)  και τους  7 ταξινομητές που προαναφέρθηκαν (εμφανίστε τα δείγματα από το train set μαζί με την ετικέτα τους στην είσοδο του κάθε ταξινομητή), χρησιμοποιώντας τις default τιμές για όλες τις παραμέτρους τους.

### 6. Βελτιστοποίηση μοντέλων με εύρεση καλύτερων υπερπαραμέτρων των ταξινομητών

```python
from sklearn.model_selection import GridSearchCV
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Initialize the classifiers
classifiers = {
    "Naive Bayes": GaussianNB(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(),
    "MLP Classifier": MLPClassifier(),
    "Support Vector Classifier": SVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}

# Define hyperparameter grids for each classifier
param_grids = {
    "Naive Bayes": {
        "var_smoothing": [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
    },
    "K-Nearest Neighbors": {
        "n_neighbors": [3, 5, 7, 9],
        "weights": ["uniform", "distance"],
        "metric": ["euclidean", "manhattan"]
    },
    "Logistic Regression": [
        {
            "penalty": ["l2"],
            "C": [0.01, 0.1, 1, 10, 100],
            "solver": ["lbfgs"]
        },
        {
            "penalty": ["l1", "l2"],
            "C": [0.01, 0.1, 1, 10, 100],
            "solver": ["liblinear"]
        },
        {
            "penalty": ["elasticnet"],
            "C": [0.01, 0.1, 1, 10, 100],
            "solver": ["saga"],
            "l1_ratio": [0.1, 0.5, 0.9]
        }
    ],
    "MLP Classifier": {
        "hidden_layer_sizes": [(50,), (100,), (50, 50)],
        "activation": ["tanh", "relu"],
        "solver": ["adam", "sgd"],
        "alpha": [0.0001, 0.001, 0.01]
    },
    "Support Vector Classifier": {
        "C": [0.1, 1, 10],
        "kernel": ["linear", "rbf", "poly"],
        "gamma": ["scale", "auto"]
    },
    "Decision Tree": {
        "criterion": ["gini", "entropy"],
        "max_depth": [None, 10, 20, 30],
        "min_samples_split": [2, 5, 10]
    },
    "Random Forest": {
        "n_estimators": [50, 100, 200],
        "criterion": ["gini", "entropy"],
        "max_depth": [None, 10, 20],
        "min_samples_split": [2, 5, 10]
    }
}

# Store the best models, hyperparameters, and training times
best_models = {}
best_params = {}
training_times = {}

# Perform GridSearchCV for each classifier
for name, clf in classifiers.items():
    print(f"\nOptimizing classifier: {name}")

    # Get the hyperparameter grid
    param_grid = param_grids[name]

    # Create the GridSearchCV object
    grid_search = GridSearchCV(
        estimator=clf,
        param_grid=param_grid,
        scoring="f1",  # Use F1-score as the evaluation metric
        cv=5,          # 5-fold cross-validation
        verbose=1,      # Display progress
        n_jobs=-1       # Parallel execution
    )

    # Measure training time
    start_time = time.time()
    grid_search.fit(X_train, y_train)
    elapsed_time = time.time() - start_time

    # Save the results
    best_models[name] = grid_search.best_estimator_
    best_params[name] = grid_search.best_params_
    training_times[name] = elapsed_time  # Store training time

    # Display the results
    print(f"Best hyperparameters for {name}: {grid_search.best_params_}")
    print(f"Best F1-score: {grid_search.best_score_:.4f}")
    print(f"Training time for {name}: {elapsed_time:.2f} seconds")
```
